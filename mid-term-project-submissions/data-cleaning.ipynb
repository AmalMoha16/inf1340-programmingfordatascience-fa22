# TABLE 1

- Extracting data
- Basic Redressing and Tidying 

```python
#Extracting the data for table 1
df_table1 = pd.read_excel("/Users/audreymedaino-tardif/Intro-Pyth/inf1340-programmingfordatascience-fa22/project data/UN_MigrantStockTotal_2015.xlsx", sheet_name="Table 1")
df_table1.columns.values.tolist()
```

```python
# Tinkering with the data => Removing the Titles without saving the removal.
removeTitles = df_table1.dropna(thresh = 3, inplace = True)
print(removeTitles)
len(df_table1.index)


# Renaming the columns
df_table1.rename(columns = {
    'Unnamed: 0' : "Sort/Order",
    'Unnamed: 1' : 'Name',
    'Unnamed: 2': 'Notes',
    'Unnamed: 3' : 'Country Code',
    'Unnamed: 4' : 'Type of data(a)',
    'Unnamed: 5' : 'T1990',
    'Unnamed: 6' : 'T1995',
    'Unnamed: 7': 'T2000',
    'Unnamed: 8' : 'T2005',
    'Unnamed: 9': 'T2010',
    'Unnamed: 10': 'T2015',
    'Unnamed: 11':'M1990',
    'Unnamed: 12':'M1995',
    'Unnamed: 13':'M2000',
    'Unnamed: 14': 'M2005',
    'Unnamed: 15': 'M2010',
    'Unnamed: 16': 'M2015',
    'Unnamed: 17':'F1990',
    'Unnamed: 18':'F1995',
    'Unnamed: 19':'F2000',
    'Unnamed: 20':'F2005', 
    'Unnamed: 21' : 'F2010',
    'Unnamed: 22': 'F2015'
     }, inplace=True)

df_table1.reset_index()


df_table1.drop(df_table1.index[0:2], inplace=True)
display(df_table1.head(20))
```

```python

# Change index to Sort/Order 
df_table1.set_index("Sort/Order")
display(df_table1.head(3))
```

```python
# Molten Dataset 
df_melted = df_table1.melt(id_vars = ["Sort/Order", 'Name', 'Notes', 'Country Code', 'Type of data(a)'], var_name="Year", value_name="# of Migrants")
display(df_melted.head(4))

```

```python
df_sep=(df_melted.assign(gender = lambda x: x.Year.str[0].astype(str), years = lambda x: x.Year.str[1:].astype(str)).drop("Year",axis=1))
df_sep.head(5)
```

```python
# Creating a missing data percentage list

for col in df_sep.columns:
    pct_missing = np.mean(df_sep[col].isnull())
    print('{} - {}%'.format(col,round(pct_missing*100)))
```

```python
# Removing Summarizing regions from table as they don't provide much information with regards to what is tabulated in the totals
df_summary = df_sep.iloc[0:5].reset_index().set_index("Sort/Order")
df_table1 = df_sep.iloc[6:].reset_index().set_index("Sort/Order")

display(df_summary)
display(df_table1)
```
